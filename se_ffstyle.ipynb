## 2. End-of-Session Cleanup (Industry Standard)

What good developers do before closing the laptop:

**Commit WIP:**
```bash
git add -A && git commit -m "WIP: description of where I left off"
```
Never leave uncommitted work overnight.

**Push to remote:** Protects against local machine failure.

**Update your tracker:** 2 minutes of session notes saves 20 minutes of "where was I?" next time.

**Clean up branches:**
```bash
git branch --merged | grep -v main | xargs git branch -d
```

**Close running servers/processes:** Avoid port conflicts next session.

**Leave a breadcrumb:** `# TODO: NEXT SESSION - ...` comment at the exact line you stopped.

## 3. Structured Approaches (For When the Project Grows)

**GitHub Issues as Sprint Items**
- One issue per sprint task, close as you go
- Use milestones for sprints — gives you a free Kanban board

**Git Stash with Messages**
```bash
git stash push -m "halfway through auth middleware"
```
When you don't want WIP commits polluting history.

**ADR (Architecture Decision Records)**
```
/docs/decisions/001-chose-neon-over-supabase.md
```
Tracks *why* you made choices — invaluable months later.

**Conventional Commits**
```
feat(auth): add JWT token refresh endpoint
fix(inventory): correct stock count on partial delivery
chore(deps): update fastapi to 0.109.0
```
Structured messages that enable automated changelogs.

**Pre-commit Hooks** — Automate quality checks: linting, formatting, type checking, secret scanning.
